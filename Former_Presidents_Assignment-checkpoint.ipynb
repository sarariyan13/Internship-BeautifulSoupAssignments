```python
page=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
page
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
former_president=soup.find('div', class_="presidentListing")
former_president.text
```

```python
former_president.text
```

```python
presidents=[]
for i in soup.find_all('div', class_="presidentListing"):
    presidents.append(i.text)
presidents
    
```

```python
import pandas as pd
df=pd.DataFrame({'Former_Presidents':presidents})
df
```
